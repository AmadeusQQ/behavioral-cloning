```python
def generate_sample():
    file = open(os.path.join(DRIVING_LOG_PATH, DRIVING_LOG_FILE), 'r')
    reader = csv.reader(file)
    
    reader.__next__()
    
    while True:
        line = reader.__next__()
        
        image = np.array(
            cv2.imread(
                os.path.join(
                    IMAGE_PATH,
                    line[0].strip('IMG/')
                )
            ),
            dtype = 'float32'
        )
        image = image.reshape(
            1,
            image.shape[0],
            image.shape[1],
            image.shape[2]
        )
        steering_angle = np.array(
            line[3], 
            dtype = 'float32')
        steering_angle = steering_angle.reshape(1, 1)
        
        yield (image, steering_angle)
    
    file.close()
       
```

```python
def generate_validation_sample():
    file = open(os.path.join(DRIVING_LOG_PATH, DRIVING_LOG_FILE), 'r')
    reader = csv.reader(file)
    reader = reversed(list(reader))
    
    while True:
        line = reader.__next__()
        
        image = np.array(
            cv2.imread(
                os.path.join(
                    IMAGE_PATH,
                    line[0].strip('IMG/')
                )
            ),
            dtype = 'float32'
        )
        image = image.reshape(
            1,
            image.shape[0],
            image.shape[1],
            image.shape[2]
        )
        steering_angle = np.array(
            line[3],
            dtype = 'float32'
        )
        steering_angle = steering_angle.reshape(1, 1)
        
        yield (image, steering_angle)
    
    file.close()
```

# Design model

Use model by NVIDIA.

Source: https://arxiv.org/pdf/1604.07316v1.pdf

```python
# Add normalization layer
# Add dropout layer
```

```python
# width = normalized_features.shape[1]
# length = normalized_features.shape[2]
# depth = normalized_features.shape[3]
```

```python
width = 160
length = 320
depth = 3
```

```python
convolution_filter = 24
kernel_size = 5
stride_size = 2

model = Sequential()
model.add(Convolution2D(
    convolution_filter,
    kernel_size,
    kernel_size,
    border_mode = 'valid',
    subsample = (stride_size, stride_size),
    input_shape = (width, length, depth)
))

convolution_filter = 36

model.add(Convolution2D(
    convolution_filter,
    kernel_size,
    kernel_size,
    border_mode = 'valid',
    subsample = (stride_size, stride_size),
    input_shape = (width, length, depth)
))

convolution_filter = 48

model.add(Convolution2D(
    convolution_filter,
    kernel_size,
    kernel_size,
    border_mode = 'valid',
    subsample = (stride_size, stride_size),
    input_shape = (width, length, depth)
))

convolution_filter = 64
kernel_size = 3

model.add(Convolution2D(
    convolution_filter,
    kernel_size,
    kernel_size,
    border_mode = 'valid',
    input_shape = (width, length, depth)
))
model.add(Convolution2D(
    convolution_filter,
    kernel_size,
    kernel_size,
    border_mode = 'valid',
    input_shape = (width, length, depth)
))
model.add(Flatten())
model.add(Dense(100))
model.add(Dense(50))
model.add(Dense(10))
model.add(Dense(1))

print('Shape:', model.output_shape)
```

# Train model

```python
# Training on a GPU is 20 times faster than CPU
```

```python
# Use transfer learning to refine a working model
```

```python
adam = Adam(lr = 0.000001)
```

```python
# model.compile(optimizer = adam, loss = 'mse')
# history = model.fit(normalized_features, labels, nb_epoch = EPOCH, validation_split = VALIDATION_SET_SIZE)
```

```python
model.compile(optimizer = adam, loss = 'mse')
history = model.fit_generator(
    generate_sample(),
    samples_per_epoch = SAMPLES_PER_EPOCH,
    nb_epoch = EPOCH,
    validation_data = generate_validation_sample(),
    nb_val_samples = VALIDATION_SET_SIZE
)
```
